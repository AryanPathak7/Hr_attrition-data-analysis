```python
sns.set(style="dark")
sns.barplot(x="Department",y="JobSatisfaction",data=data,ci=None,hue="Gender",errcolor="g")
plt.show()
```

**From this plot we conclude that Female's employee are more satisfied with their job in "Human Resources" department and Male's employee are more satisfied in"Research & Development" Department.**

```python
fig,ax=plt.subplots(ncols=2,figsize=(14,5))
var=sns.boxplot(x='Attrition',y="Age",data=data,ax=ax[0])
var.set_title("Comparison of AGE vs Attrition")
var1=sns.distplot(data.Age,kde=False,ax=ax[1])
var1.set_ylabel("Frequency")
var1.set_title("Distribution of Age")
plt.show()
```

**-> Median age of employee is between 30-40yrs.**

**->From the boxplot we consider that major employee who left the company are below 40 Yrs and those who didn't left the company are of age 35-40Yrs.**

```python
fig,ax=plt.subplots(ncols=2,figsize=(14,5))
var=sns.barplot(x='Attrition',y="DailyRate",data=data,ax=ax[0],hue="OverTime",legend="auto",palette="magma")
var.legend(title='overtime',loc= "center")
var.set_title("Comparison of DailyRate vs Attrition")
var1=sns.distplot(data.DailyRate,kde=False,ax=ax[1])
var1.set_ylabel("Frequency")
var1.set_title("Distribution of DailyRate")
plt.show()
```

**-> Average daily rate of employee is between 600to700.**

**->On the basis of daily rate the person who work overtime and have daily wages upto 800 & above are less prone to attrition and the employee who have wages lower than 800 and do overtime are more prone to attrition.**

```python
fig,ax=plt.subplots(ncols=2,figsize=(14,5))
var=sns.barplot(x='Attrition',y="MonthlyIncome",data=data,ci=None,ax=ax[0],hue="MaritalStatus",legend="auto",palette="cubehelix")
var.legend(title='MartialStatus',loc= "center")
var.set_title("Comparison of MonthlyIncome vs Attrition")
var1=sns.distplot(data.MonthlyRate,kde=False,ax=ax[1])
var1.set_ylabel("Frequency")
var1.set_title("Distribution of MonthlyIncome")
plt.show()
```

**=>From this we conclude that most of the married employee and divorced employee with lower wages has left the company and divorced employee with higher salary are more focused towards their working responsbility.**

```python
plt.figsize=(14,6)
var=sns.barplot(x='Educated',y="left",data=data,palette="viridis",ci=None)
var.set_title("Attrition by Education_Level")
var.set_xlabel("Education_level")
var.set_ylabel("Attrition")
var.grid(True)
plt.show()
```

```python
fig = plt.figure()
ax1 =fig.add_subplot(221)
ax2 =fig.add_subplot(222)
ax3 =fig.add_subplot(223)
ax4 =fig.add_subplot(224)

labels = 'Low','Medium','High','Very High'

data['JobSatisfaction'].astype(str).value_counts().plot(kind='pie',figsize=(14,6),autopct='%1.1f%%',labels=None,ax=ax1)
ax1.set_title('Rating of Job Satisfaction by Employee')
fig.legend(labels=labels,loc='center')

data['EnvironmentSatisfaction'].astype(str).value_counts().plot(kind='pie',figsize=(14,6),autopct='%1.1f%%',labels=None,ax=ax2)
ax2.set_title('Rating of Environment Satisfaction by Employee')


data['RelationshipSatisfaction'].astype(str).value_counts().plot(kind='pie',figsize=(14,6),autopct='%1.1f%%',labels=None,ax=ax3)
ax3.set_title('Rating of Relationship Satisfaction by Employee')


data['JobInvolvement'].astype(str).value_counts().plot(kind='pie',figsize=(14,6),autopct='%1.1f%%',ax=ax4,labels=None)
ax4.set_title('Rating of Job Involvement by Employee')
plt.show()
```

**From the subplots, we can infer more than 60% of the employee are:**

->Not Satisfied in their Job

->Not Satisfied with their working Environment

->Not Satisfied in their Realtionship maybe due to lower wages.

->Not Getting focused in their job Role

```python
plt.figsize=(14,6)
props = data.groupby("BusinessTravel")["Attrition"].value_counts(normalize=False).unstack()
props.plot(kind="bar",alpha=1)
plt.title("Business Travel vs Attrition")
plt.ylabel("Number of Employee")
plt.show()
```

**From above data it is clear that Employee who travel rarely have more attriton rate followed by Employees who travel frequently**

-> Best way to reduce this attrition is to conduct monthly survey and to assign travel according to Employees' business travel interest

```python

yrr=data.groupby("YearsWithCurrManager")['Attrition'].value_counts(normalize=False).unstack()
yrr.plot(kind="bar",stacked="False",figsize=(15,5))
plt.ylabel("No. of Employees")
plt.show()
```

**It is clear from above plot that Employees who are less experienced are not happy with the relation with Manager.It is important that the manager communicate with the employee from the starting itself trying to understand them soon to reduce the Attrition rate**
